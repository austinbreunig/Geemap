```python
import geemap 
import ee
```

```python
m = geemap.Map(center=[40, -100], zoom=4)
m
```

```python
region = m.draw_last_feature
roi = region.geometry()
```

```python
print(roi.getInfo())
```

```python
landsat_ts = geemap.landsat_timeseries(
    roi=roi, start_year=1984, end_year=2020, start_date='06-01', end_date='09-01'
)
```

```python
layer_names = ['Landsat ' + str(year) for year in range(1984, 2020)]
print(layer_names)
```

```python
image = ee.Image('USGS/SRTMGL1_003')
vis_params = {
  'min': 0,
  'max': 4000,
  'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']
  }
m.addLayer(image, vis_params, 'SRTM DEM', True, 0.5)
```
